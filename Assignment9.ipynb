2. Write a program which contains one function ChkGreater() that accepts two numbers and prints the greater number.
Input: 10 20
Output: 20 is greater

```python
def ChkGreater(a, b):
    if a > b:
        return a
    else :
        return b
    
def main():
    n1 = int(input("Enter 1st number: "))
    n2 = int(input("Enter 2nd Number: "))

    ans = ChkGreater(n1, n2)

    print(ans, "is greater")

if __name__ == "__main__":
    main()
    
```

3. Write a program which accepts 1 number and print square of that number.
Input: 5
Output: 25

```python
number = int(input("Enter a number : "))

sq = number * number

print(sq)
```

```python
def sq(number):
    return number * number

def main():
    no = int(input("Enter a number : "))
    print(sq(no))

if __name__ == "__main__":
    main()
```

5. Write a program which accepts one number and checks whether it is Divisible by 3 and 5. 
Input: 15 
Output: Divisible by 3 and 5

```python
no = int(input("Enter a number : "))

if no % 3 == 0 and no % 5 == 0:
    print("Divisible by 3 and 5")

else:
    print("Not Divisible by 3 and 5")
```

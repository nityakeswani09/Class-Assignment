1. Write a program which accepts one number and prints multiplication table of that number.

```python
num = int(input("Enter a number : "))
for i in range(1, 11):
    print(num*i)
```

2. Write a program which accepts one number and print sum of first N natural numbers. 
Input: 5
Output: 15

```python
n = int(input("Enter number : "))
total = 0
for i in range(1, n+1):
    total = total + i
print(total)
```

3. Write a program which accepts one number and prints factorial of that number.
Input: 5
Output: 120

```python
n = int(input("Enter number : "))
factorial = 1
for i in range(1, n+1):
        factorial = factorial * i

print("Factorial is : ", factorial)
```

4. Write a program which accepts one number and prints even numbers till that number.
Input:  10
Output: 2 4 6 8 10

```python
num = int(input("Enter a number : "))
for i in range(2, num+1, 2):
        print(i)
```

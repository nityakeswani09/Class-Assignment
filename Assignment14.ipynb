1. Write a lambda function which accepts one number and return square of that number

```python
Square = lambda no : no* no
def main():
    Data = int(input("Enter number : "))
    print("Square is : ", Square(Data))


if __name__ == "__main__":
    main()
```

2. Write a Lamda Function which accepts one number and returns cube of that number.

```python
Cube = lambda no : no **3
def main():
    no = int(input("Enter number : "))
    print("Cube is : ", Cube(no))

if __name__ =="__main__":
    main()
```

3. Write a Lambda function which accepts two numbers and return maximum number.

```python
Max = lambda no1, no2 : no1 > no2
def main():
    no1 = int(input("Enter number : "))
    no2 = int(input("Enter number : "))

    if Max(no1, no2) == True:

        print("Maximum number is : ", no1)

    else:
        print("Maximum number is : ", no2)

if __name__ == "__main__":
    main()
```

4. Write a Lambda Function which accepts two numbers and return minimum number

```python
Min = lambda no1, no2 : no1 < no2

def main():
    no1 = int(input("Enter 1st number : "))
    no2 = int(input("Enter 2nd number : "))

    if Min(no1, no2) == True :
        print("Minimum number is :", no1)

    else : 
        print(f"Minimum number is : {no2}")

if __name__ == "__main__":
    main()
```

5. Write a lambda function which accepts 1 no and returns True if number is even  otherwise False.

6. Write a lambda function which accepts one number and returns True if number is odd otherwise False.

```python
Odd = lambda no : no % 2 != 0

def main():
    n = int(input("Enter number : "))

    if Odd(n) == True:
        print("True.")

    else:
        print("False.")

if __name__ == "__main__":
    main()
```

8. Write a Lambda Function which accepts 2 numbers and return addition.

```python
Add = lambda num1, num2 : num1 + num2

def main():
    n1 = int(input("Enter 1st number : "))
    n2 = int(input("Enter 2nd number : "))

    print("Addition is : ", Add(n1, n2))

if __name__ == "__main__":
    main()
```

9. Write a lambda function which accepts two numbers and return Multiplication.

```python
Mul = lambda no1, no2 : no1 * no2

def main():
    number1 = int(input("Enter 1st Number : "))
    number2 = int(input("Enter 2nd Number : "))
    print(f"Multiplication of {number1} into {number2} is {Mul(number1, number2)}")

if __name__ == "__main__":
    main()
```

10. Write a lambda function which accepts three numbers and return largest number.

```python
Largest = lambda no1, no2, no3 : no1 if no1 > no2 and no1 > no3 else (no2 if no2 > no3 else no3)

def main():
    no1 = int(input("Enter 1st number : "))
    
    no2 = int(input("Enter 2nd number : "))
    
    no3 = int(input("Enter 3rd number : "))

    Ans = Largest(no1, no2, no3)
    print("Largest number is : ", Ans)

if __name__ == "__main__":
    main()


```

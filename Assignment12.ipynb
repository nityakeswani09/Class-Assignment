1. Write a program which accepts 1 character and checks whether it is vowel or consonant.
Input: a
Output: Vowel

```python
ch = input("Enter a character: ")

if ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u' or ch == 'A' or ch == 'E' or ch == 'I' or ch == 'O' or ch == 'U':
    print("Vowel")
else:
    print("Consonant")
```

2. Write a program which accepts one number and print its factors.
Input : 12
Output : 1 2 3 4 6 12

```python
num = int(input("Enter a number: "))

for i in range(1, num + 1):
    if num % i == 0:
        print(i, end=" ")
```

3. Write a program which accepts two numbers and prints addition, sybtraction, multiplication and division.

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("Addition =", num1 + num2)
print("Subtraction =", num1 - num2)
print("Multiplication =", num1 * num2)
print("Division =", num1 / num2)
```

4. Write a program which accepts one number and prints that many numbers starting from 1.
Input : 5
Output : 1 2 3 4 5

```python
num = int(input("Enter a number: "))

for i in range(1, num + 1):
    print(i)
```

5. Write a program which accepts one number and prints that many numbers in reverse order.
Input : 5
Output :  5 4 3 2 1 

```python
num = int(input("Enter a number: "))

for i in range(num, 0, -1):
    print(i, end=" ")
```

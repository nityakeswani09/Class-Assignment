1. Write a program which accepts one number and checks whether it is prime or not.
Input : 11
Output :  Prime Number

```python
num = int(input("Enter number: "))

is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break                   #Not neccessary but reduces computers task
    
if is_prime:
    print("Prime")
else:
    print("Not Prime")
```

2. Write a program which accepts one number and print count of digits in that number.
Input : 7521
Output: 4

```python
n = int(input("Enter number to count digits: "))

if n == 0:
    count = 1
else:
    count = 0
    while n > 0:
        count = count + 1
        n = n // 10

print("Number of digits =", count)
```

3. Write a program which accepts one number and prints sum of digits
Input : 123
Output: 6

```python
num = int(input("Enter a number: "))

sum = 0

while num > 0:
    digit = num % 10
    sum = sum + digit
    num = num // 10

print("Sum of digits =", sum)
```

4. Write a program which accepts one number and print reverse of that number
Input  : 123
Output : 321

```python
num = int(input("Enter a number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reverse =", reverse)
```

5. Write a program which accepts one number and checks whether it is palindrome or not.
Input : 121
Output : Palindrome

```python
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```
